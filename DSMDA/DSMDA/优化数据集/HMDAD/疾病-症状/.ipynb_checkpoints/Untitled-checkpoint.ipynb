```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Excel
data = pd.read_excel('disease_with_symptoms.xlsx')

# 
data = data.dropna(subset=['Symptoms'])

# 
diseases = data['Disease'].unique()
symptoms = data['Symptoms'].str.split(',').explode().unique()

# 
symptoms = [symptom.strip() for symptom in symptoms]

# 
disease_to_index = {disease: idx for idx, disease in enumerate(diseases)}
symptom_to_index = {symptom: idx for idx, symptom in enumerate(symptoms)}

# A
A = np.zeros((len(diseases), len(symptoms)), dtype=int)

# A
for _, row in data.iterrows():
    disease_idx = disease_to_index[row['Disease']]
    symptom_list = [symptom.strip() for symptom in row['Symptoms'].split(',')]
    for symptom in symptom_list:
        if symptom in symptom_to_index:
            symptom_idx = symptom_to_index[symptom]
            A[disease_idx, symptom_idx] = 1

# D
D = np.diag(A.sum(axis=1))

# D
D_inv_sqrt = np.linalg.inv(np.sqrt(D))

# S
S = D_inv_sqrt @ A @ A.T @ D_inv_sqrt

# CSV
pd.DataFrame(S, index=diseases, columns=diseases).to_csv('disease_similarity_matrix.csv')

# 
plt.figure(figsize=(10, 8))
sns.heatmap(S, cmap='viridis', xticklabels=diseases, yticklabels=diseases, annot=False, cbar=True)
plt.title('Disease Similarity Matrix')
plt.xlabel('Disease Names')
plt.ylabel('Disease Names')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
